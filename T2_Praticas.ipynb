```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
inflacao = pd.read_csv ('https://raw.githubusercontent.com/eduardoluisvieira/mackenzie/main/STP-inflacao.csv', delimiter=';',decimal=',')
```

```python
inflacao.head
```

```python
plt.figure(figsize=(20,5), dpi = 80)
IPCA_Servicos = inflacao[ inflacao.IPCA == 'IPCA' ]
IPCA_Geral = inflacao[ inflacao.IPCA_Servico == 'IPCA Servico' ]

plt.subplot(1,2,1)
#plt.plot(inflacao['Data'],inflacao['IPCA'],'-')
plt.plot(inflacao.Data, inflacao.IPCA, label='IPCA')
plt.plot(inflacao.Data, inflacao.IPCA_Servico, label='IPCA_Servico')
plt.title('Inflação IPCA x IPCA Serviços',fontsize=14,weight='bold')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Média do IPCA:
IPCA_Series = inflacao['IPCA']
print (f'Média do IPCA:{IPCA_Series.mean():.2f}%')
```

```python
# Média do IPCA de Serviços
IPCAServico_Series = inflacao['IPCA_Servico']
print (f'Média do IPCA de Serviços:{IPCAServico_Series.mean():.2f}%')
```

```python
# As cinco menores taxas de inflação
inflacao.nsmallest(5,'IPCA')
```

```python
# As cinco maiores taxas de inflação
inflacao.nlargest(5,'IPCA')
```

```python
# IPCA > 10%
IPCA_10 = inflacao[inflacao.IPCA > 10.0][['IPCA', 'IPCA_Servico']]
IPCA_10
```
