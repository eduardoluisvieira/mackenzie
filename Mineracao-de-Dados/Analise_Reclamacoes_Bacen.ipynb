```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
financeiras = pd.read_csv('https://raw.githubusercontent.com/eduardoluisvieira/mackenzie/main/Bancos.csv', delimiter=';',decimal=',', encoding = 'latin1')
```

```python
financeiras.head()
```

```python
df_financeiras = financeiras[['Instituição financeira','Índice','Quantidade total de reclamações','Quantidade total de clientes  CCS e SCR']]
```

```python
df_financeiras
```

```python
# Bancos com menos reclamações
df_financeiras.nsmallest(3,'Quantidade total de reclamações')
```

```python
# Top 3 bancos com mais reclamações
top3 = df_financeiras.nlargest(3,'Quantidade total de reclamações')
top3
```

```python
total_reclamacoes = df_financeiras['Quantidade total de reclamações']
total_reclamacoes
```

```python
plt.figure(figsize=(30,6))
plt.subplot(1,2,1)
sns.barplot(x=top3['Instituição financeira'],y=top3['Quantidade total de reclamações'],color='r',label='reclamações')
plt.title('Quantidade de Reclamações por Instituição Financeira',fontsize=12,weight='bold')
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```
